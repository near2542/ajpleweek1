```python
import pandas as pd
url="https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv"
c=pd.read_csv(url)
c.sample(3)
```

```python
unnecessary_data = ['hiv','size','utmet','vgmet','surgery','pchemo']
c = c.drop(unnecessary_data,1)
```

```python
c
```

```python
c['ages'] = c['age'] > 20
c
```

```python
c['ages'] = c['ages'].replace(True,'old')
c['ages'] = c['ages'].replace(False,'young')
c
```

```python
c.drop(['age','ages'],1) #drop cuz all of them are >20
```

```python
c['pmmet'].unique()
```

```python
c = c.dropna(0,'any') #drop unavailable row
c['RHlvsi'].unique()
```

```python
import numpy as np

pmmet0 = c[(c['pmmet']== 0.0)]
pmmet0 = pmmet0.drop('ages',1)
pmmet0
```

```python
pmmet1 = c[(c['pmmet']== 1.0)]
pmmet1 = pmmet1.drop('ages',1)
pmmet1
```

```python
pmmet0Sample = pmmet0.sample(10)
pmmet0Sample
```

```python
pmmet1Sample = pmmet1.sample(10)
pmmet1Sample
```

```python
import matplotlib.pyplot as plt

pm0Patient = pmmet0['disease'].count()
pm1Patient = pmmet1['disease'].count()


fig = plt.figure(1, figsize=(5,6))
pm0DiseasePercentage = pmmet0['disease'][(pmmet0['disease']==1)].count()/pm0Patient*100
pm1DiseasePercentage = pmmet1['disease'][(pmmet1['disease']==1)].count()/pm1Patient*100

#Disease Rate comparision

plt.title('Patients Disease Rate')
plt.ylabel('Percentage')
plt.bar(['pmmet0','pmmet1'] , [pm0DiseasePercentage,pm1DiseasePercentage])

```

```python
vgmargin_sorted_rate = [
                    c['vgmargin'][(c['vgmargin']==0)].count()/c['vgmargin'].count()*100,
                    c['vgmargin'][(c['vgmargin']==1)].count()/c['vgmargin'].count()*100,
                    c['vgmargin'][(c['vgmargin']==2)].count()/c['vgmargin'].count()*100,
                    ]
vgmargin_criteria = ['0','1','2']

plt.title('vgmargin rate on all patient')
plt.ylabel('Percentage')
plt.bar(vgmargin_criteria , vgmargin_sorted_rate)
```

```python
vgmargin_sorted_rate_of_pmmet0 = [
                    pmmet0['vgmargin'][(pmmet0['vgmargin']==0)].count()/pmmet0['vgmargin'].count()*100,
                    pmmet0['vgmargin'][(pmmet0['vgmargin']==1)].count()/pmmet0['vgmargin'].count()*100,
                    pmmet0['vgmargin'][(pmmet0['vgmargin']==2)].count()/pmmet0['vgmargin'].count()*100,
                    ]
vgmargin_sorted_rate_of_pmmet1 = [
                    pmmet1['vgmargin'][(pmmet1['vgmargin']==0)].count()/pmmet1['vgmargin'].count()*100,
                    pmmet1['vgmargin'][(pmmet1['vgmargin']==1)].count()/pmmet1['vgmargin'].count()*100,
                    pmmet1['vgmargin'][(pmmet1['vgmargin']==2)].count()/pmmet1['vgmargin'].count()*100,
                    ]

fig = plt.figure(1, figsize=(15,6))

plt.subplot(1,2,1)
plt.title('pmmet0 vgmargin percentage')
plt.ylabel('percentage')
plt.bar(vgmargin_criteria, vgmargin_sorted_rate_of_pmmet0)

plt.subplot(1,2,2)
plt.title('pmmet1 vgmargin percentage')
plt.ylabel('percentage')
plt.bar(vgmargin_criteria, vgmargin_sorted_rate_of_pmmet1)
```
