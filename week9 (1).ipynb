```python
y=df_clean[0].values
y
```

```python
X=df_clean.drop(0,axis=1).values
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
fs = SelectKBest(chi2, k=600)
fs.fit(X_train, y_train)
X_test_norm = fs.transform(X_test)
X_train_norm = fs.transform(X_train)
X_train.shape
```

```python
X_train_norm.shape
```

```python
X_test_norm.shape
```

```python
from sklearn.ensemble import StackingClassifier

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from catboost import CatBoostClassifier

from sklearn.naive_bayes import GaussianNB


clf1 = CatBoostClassifier(random_state=1,verbose=False,iterations=50,learning_rate=1)

clf2 = CatBoostClassifier(random_state=1,verbose=False,iterations=50,learning_rate=1)

clf3 = CatBoostClassifier(random_state=1,verbose=False,iterations=50,learning_rate=1)

clf4 = CatBoostClassifier(random_state=1,verbose=False,iterations=50,learning_rate=1)

clf5 = GaussianNB()

eclf = StackingClassifier(

estimators=[('cb1', clf1), ('cb2', clf2), ('cb3', clf3) , ('cb4',clf4) , ('gnb',clf5)],

stack_method = 'predict_proba',

final_estimator= CatBoostClassifier(random_state=1,iterations=51))

params = {'cb1__depth': [1,5], 'cb2__depth': [1,5] , 'cb3__depth':[1,5] ,'cb4__depth':[1,5] ,}

grid = GridSearchCV(estimator=eclf, param_grid=params, cv=5)

grid = grid.fit(X_train_norm,y_train)

print("Best params : " + str(grid.best_params_))
print("10CV accuracy : "+str(grid.best_score_*100))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
y_predict = grid.predict(X_test_norm)

target_names = ['negative', 'positive']
sum(y_test == y_predict)/len(y_test)
```
