```python
fig = plt.figure(1, figsize=(15,6))

# bar chart example
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300]

plt.subplot(1,2,1)
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem)

# pie chart example
labels = 'fabiyamada', 'jedigeiss', 'scipio', 'stoodkev'
perc = [steem[0]/sum(steem), # calculate percentages from the amount of steem
        steem[1]/sum(steem), 
        steem[2]/sum(steem), 
        steem[3]/sum(steem)]

explode = (0, 0, 0.1, 0)     # "explode" the 3rd slice ('Scipio')

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')            # Make sure the pie is drawn as a circle, X and Y axes equally sized
plt.pie(perc, labels=labels, autopct='%1.2f%%', 
        startangle=90, explode=explode)

plt.show()
```

```python
titanics = np.delete(titanics,2,1)
titanics[0,:]
```

```python
passenger_class = np.unique(titanics[1:,0])
passenger_class
```

```python
passenger_class_value = [sum(titanics[:,0]==passenger_class[0]),
                         sum(titanics[:,0]==passenger_class[1]),
                         sum(titanics[:,0]==passenger_class[2]) ]
passenger_class_value
```

```python
plt.bar(passenger_class, passenger_class_value)
```

```python
passenger_sex= np.unique(titanics[1:,1])
passenger_sex
passanger_sex_value = [sum(titanics[:,1] == passenger_sex[0]),
                       sum(titanics[:,1] == passenger_sex[1])]
passanger_sex_value
plt.subplot(1,2,1)
plt.bar(passenger_sex, passanger_sex_value)

passenger_survive= np.unique(titanics[1:,4])
passenger_survive

passenger_survive_value = [sum(titanics[:,4] == passenger_survive[0]),
                       sum(titanics[:,4] == passenger_survive[1])]

plt.subplot(1,2,2)
plt.bar(passenger_survive,passenger_survive_value)
plt.show()
    
```
